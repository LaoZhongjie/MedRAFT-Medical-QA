```python
import json
from collections import defaultdict

# 1. è¯»å–åŸå§‹ JSON æ–‡ä»¶
with open("../data/diseases.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# 2. æŒ‰ category åˆ†ç»„
grouped = defaultdict(list)
for item in data:
    category = item["category"]
    disease = item["disease_name"]
    grouped[category].append(disease)

# 3. è¾“å‡ºæ¯ä¸ªç±»åˆ«åŒ…å«å¤šå°‘ç§ç–¾ç—…
print("ğŸ“Š å„ç±»åˆ«ç–¾ç—…æ•°é‡ï¼š")
total_count = 0
for category, diseases in grouped.items():
    count = len(diseases)
    total_count += count
    print(f" - {category}: {count} ç§")

# 4. è¾“å‡ºæ€»çš„ç–¾ç—…æ•°é‡
print(f"\nâœ… æ€»ç–¾ç—…æ•°é‡: {total_count} ç§")

# 5. å¦‚æœæƒ³ä¿å­˜ç»Ÿè®¡ç»“æœä¸º JSON æ–‡ä»¶
stats = {
    "total_disease_count": total_count,
    "categories": {cat: len(diseases) for cat, diseases in grouped.items()}
}

with open("disease_stats.json", "w", encoding="utf-8") as f:
    json.dump(stats, f, ensure_ascii=False, indent=4)

print("\nğŸ“ ç»Ÿè®¡ç»“æœå·²ä¿å­˜åˆ° disease_stats.json")
```
